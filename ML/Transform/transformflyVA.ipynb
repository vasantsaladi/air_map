```python

# Load the dataset
df = pd.read_csv('/Users/vasantsaladi/Documents/GitHub/air_map/ML/Data/Virginiafly.csv')  # This is the correct filename
df.head()
```

```python
# Transform Data Frame
df = df.drop(['scientific_name', 'place_guess','place_town_name', 'time_zone', 'observed_on_string','time_observed_at'], axis=1)
#dropped the columns that are not needed for the model
df.head()
```

```python
mean_positional_accuracy = 3422.3690676945316
print(mean_positional_accuracy)



```

```python

rows_before = len(df)

# Filter the dataframe
df = df[df['positional_accuracy'] < mean_positional_accuracy]

# Calculate the number of rows after filtering
rows_after = len(df)

# Calculate and print the number of rows removed
rows_removed = rows_before - rows_after
print(f"Number of rows removed: {rows_removed}")
print(f"Number of rows after filtering: {rows_after}")

df.head()








```

```python
# Export the filtered dataframe to a new CSV file
output_path = '/Users/vasantsaladi/Documents/GitHub/air_map/ML/Data/marylandflyml.csv'
df.to_csv(output_path, index=False)
print(f"Filtered data exported to: {output_path}")
```
